Answer: I decided to deal with the missing values using a combination of a flag variable and imputation. The variable with the missing values was the one that described how many days had been since a customer last made a purchase. This means, the value was missing likely because it was a new customer, so this record cannot be deleted or imputated. First, I flagged the values, then I made a subset to verify that the values had unique User IDs to make sure they were new or unique customers, then I made a new column where I copied the old one and imputated '0' as the new value for NaN. The reason why I did not simply imputate directly in the old column is because I did not check whether the subset of User IDs repeated in the dataframe as a whole.
