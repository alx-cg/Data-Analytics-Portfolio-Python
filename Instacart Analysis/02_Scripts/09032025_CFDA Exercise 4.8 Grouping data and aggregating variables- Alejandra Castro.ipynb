4. Aggregating data

Single aggregation

```python
#Single aggregation to calculate mean of 'order_number' by 'department_id'
df.groupby('department_id').agg({'order_number': ['mean']})
#here groupby is being assigned to df, this creates a pandas object for dept._id, agg() is applied onto 
#this object and will return the mean of the column
```

```python
#Some aggregations can be done without the agg() function.
df.groupby('department_id')['order_number'].mean()
```

```python
#Alternative notation: dot notation
df.groupby('department_id').order_number.mean()
```

```python
#Reasons why not to use dot notation: 1) Visuals: Square brackets are easier to see. 2) Default roles: Square brackets do not have another role other than indexing, but dots do (applying panda functions for example)
```

Multiple aggregations

```python
#Calculation mean, min, and max
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

Aggregating data with transform()

```python
#transform() creates a new column containing the maximum frequency of the 'order_nummber' column and then, using loc(), it creates a second column containing a flag
```

```python
#Criteria: 
#If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
#If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
#If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”
```

```python
#Steps: 
#1. Split the data into groups based on the “user_id” column. (groupby)
#2. Apply the transform() function on the “order_number” column to generate the maximum orders for each user. (transform)
#3. Create a new column, “max_order,” into which you’ll place the results of your aggregation. (loc)
```

```python
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Operation check
ords_prods_merged.head(10)
```

```python
#Alternate check: first 100 records
ords_prods_merged.head(100)
```

```python
#Jupyter normally prevents all records from showing ('...' break), to override: 
#pd.options.display.max_rows= None
```

```python
#Deriving columns with loc()
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
